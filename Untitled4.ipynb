```python
files = ('airports', 'planes', 'airlines', 'weather')

dfs = {
    file: pd.read_csv(f'{file}.csv')
    for file in files
}

dfs['flights'] = pd.read_csv('flights.csv', skiprows=54, parse_dates = ['time_hour'])
```
