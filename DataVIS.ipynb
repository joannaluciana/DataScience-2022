```python
births = pd.read_csv('https://raw.githubusercontent.com/jakevdp/data-CDCbirths/master/births.csv')
```

```python
births.head()
```

```python
births['decade'] = 10 * (births['year'] //10)
```

```python
births.head()
```

```python
births.pivot_table('births', index ='decade', columns='gender', aggfunc ='sum').plot(kind='bar')
```

```python
births.pivot_table('births', index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel('total births per year')
```

```python
quartiles = np.percentile(births['births'], [25,50,75])
mu = quartiles[1]
sig = 0.74 * (quartiles[2] - quartiles[0])
```

```python
births = births.query('(births > @mu - 5*@sig) & (births < @mu + 5*@sig)')
births.describe()
```

```python
10000 * births.year+100 * births.month + births.day, format = '%Y%m%d'
```

```python
births.index = pd.to_datetime(10000 * births.year+100 * births.month + births.day, format = '%Y%m%d')
```

```python
births.head()
```

```python
births['dayofweek'] = births.index.dayofweek
```

```python
births.head()
```

```python
births['dayofweek'] = births.index.strftime('%a')
births.head()
```

```python
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
births.pivot_table('births', index='dayofweek', columns = 'decade', aggfunc = 'mean'). loc[weekdays].plot()
```

```python
births_by_date=births.pivot_table('births', [births.index.month,births.index.day])
births_by_date
```

```python
from datetime import datetime

births_by_date.index = [datetime(2012, month, day)
                        for (month, day) in births_by_date.index]

births_by_date

fig, ax = plt.subplots(figsize=(12, 4))
births_by_date.plot(ax=ax)

style = dict(size=10, color='gray')

ax.text('2012-1-1', 3950, "New Year's Day", **style)
ax.text('2012-7-4', 4250, "Independence Day", ha='center', **style)
ax.text('2012-9-4', 4850, "Labor Day", ha='center', **style)
ax.text('2012-10-31', 4600, "Halloween", ha='right', **style)
ax.text('2012-11-25', 4450, "Thanksgiving", ha='center', **style)
ax.text('2012-12-25', 3850, "Christmas ", ha='right', **style)

ax.set(title='USA births by day of year (1968-2008)',
       ylabel='avg daily births')


ax.xaxis.set_major_locator(mpl.dates.MonthLocator())
ax.xaxis.set_minor_locator(mpl.dates.MonthLocator(bymonthday=15))
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.xaxis.set_minor_formatter(mpl.dates.DateFormatter('%b'))
```

```python
marathon_url = 'https://raw.githubusercontent.com/jakevdp/marathon-data/master/marathon-data.csv'
```

```python
marathon_df=pd.read_csv(marathon_url)
```

```python
marathon_df.head()
```

```python
marathon_df.dtypes
```

```python
from datetime import timedelta

def convert_time(s):
    h, m, s = map(int, s.split(':'))
    return timedelta(hours=h, minutes=m, seconds=s)

marathon_df = pd.read_csv(marathon_url,
                          converters={
                              'split': convert_time,
                              'final': convert_time
                          })
    
```

```python
marathon_df.dtypes
```

```python
marathon_df.head()
```

```python
marathon_df['split_sec'] = marathon_df['split'].dt.seconds
```

```python
marathon_df['final_sec'] = marathon_df['final'].dt.seconds
```

```python
marathon_df.head()
```

```python
with sns.axes_style('white'):
    g = sns.jointplot(x=marathon_df['split_sec'],
                      y=marathon_df['final_sec'],
                      kind='hex')
    g.ax_joint.plot(np.linspace(4000, 16000),
                    np.linspace(8000, 32000), ':k')
    
```

```python
marathon_df['split_frac'] = (
    1 - 2 * marathon_df['split_sec']/marathon_df['final_sec'])
```

```python
marathon_df.head()
```

```python
sns.distplot(marathon_df['split_frac'], kde=False);
plt.axvline(0, color='k', linestyle = '--')
```

```python
sum(marathon_df.split_frac < 0)
```

```python
g = sns.PairGrid(marathon_df, vars = ['age', 'split_sec', 'final_sec', 'split_frac'], hue = 'gender', palette = 'RdBu_r')
g.map(plt.scatter, alpha=0.5)
g.add_legend();
```

```python
sns.violinplot(
    x=marathon_df['gender'],
    y=marathon_df['split_frac'],
    palette=['lightblue', 'lightpink'])
```
