```python
x = np.linspace(0,10,100)
plt.figure()
plt.subplot(2,1,1) # wiersze, kolumny, nr panel 
plt.plot(x,np.sin(x))
plt.subplot(2,1,2) # wiersze, kolumny, nr panel 
plt.plot(x,np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x,np.sin(x))
ax[1].plot(x,np.cos(x)); # srednik bo inaczej widac opis
```

```python
plt.plot(x,np.sin(x), color = 'pink');
plt.plot(x,np.sin(x-1), color = 'g');
plt.plot(x,np.sin(x-2), color = '#FFDD44');
plt.plot(x,np.sin(x-3), color = 'chartreuse');

```

```python
plt.plot(x,x, linestyle = 'solid');
plt.plot(x,x+1, linestyle = '-');
plt.plot(x,x+2, linestyle = '--');
plt.plot(x,x+3, linestyle = ':');
```
